## 재귀함수... 수열의 귀납적 정의 떠올리기

$$a_n = n \times a_{n-1} \ (n \ge 2)$$

```python
def factorial1(n):
    output = 1
    for i in range(1, n+1):
        output *= i
    return output

print(factorial1(4))
print(factorial1(5))


def factorial2(n):
    if n == 0:
        return 1
    else:
        return n * factorial2(n-1)

print(factorial2(4))
print(factorial2(5))
```

$$a_n = a_{n-1} + a_{n-2} \  (n \ge 3)$$

```python
count = 0
def fibonacci1(n):
  global count   # 전역 변수 count 사용 선언
  if n == 1:
    count+=1
    print("fibonacci1(" + "{}".format(n) + ") 연산, " + "{}".format(count) + "번째 연산")
    return 1
  if n == 2:
    count+=1
    print("fibonacci1(" + "{}".format(n) + ") 연산, " + "{}".format(count) + "번째 연산")
    return 1
  else:
    count+=1
    print("fibonacci1(" + "{}".format(n) + ") 연산, " + "{}".format(count) + "번째 연산")
    return fibonacci1(n-1) + fibonacci1(n-2)

print(fibonacci1(10))

# 자바와 달리 count += 1이 값을 반환하지 않는다.
# 비동기적 함수 호출이더라도 전역 변수를 활용해 값 공유 가능
# count 포함 로직을 함수로 만들면 재사용 가능
# 파이썬은 함수 내부에서 함수 외부에 있는 변수를 참조(변수에 접근)하지 못한다.
# "외부 영역에서 재정의됨/ 정의되지 않은 변수임" 오류에 대한 처리 -> global
```
