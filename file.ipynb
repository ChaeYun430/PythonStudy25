```python
# CSV : Comma Seperated Values
# 한 줄에 하나의 데이터를 나타내며, 각각의 줄은 쉼표를 사용해 데이터를 구분함.

# 파일을 한 줄씩 읽어 들여 처리
import random
korean = list('가나다라마바사아자차카타파하')
with open('file4.txt', 'a') as file4:
  for i in range(1000):
    name = random.choice(korean) + random.choice(korean)
    height = random.randrange(150, 200)
    weight = random.randrange(40, 100)
    file4.write('{}, {}, {}\n'.format(name, height, weight))
```

```python
with open("file4.txt", 'r') as file4:
  for line in file4:
    (name, height, weight) = line.strip().split(', ')
    if (not name) or (not height) or (not weight):
      continue # 비어있으면 스킵
    bmi = int(weight) / ((int(height) / 100) ** 2)
    print('{}, {}, {}, {}'.format(name, height, weight, bmi), sep = '\t')
```
