```python
# 요소를 하나만 가지는 튜플의 선언
tuple1 = (1234,)
print(tuple1)
print(type(tuple1))
print(tuple1[0])

# dictionary처럼 사용하는 list & tuple
[a, b] = ['A', 'B']
(c, d) = ('C', 'D')
print(a,b,c,d, sep='\t')

# 괄호를 생략한 tuple
tuple3 = 10, 20, 30, 40
for val in tuple3:
    print(val, end='\t')

a, b, c = 11, 22, 33 # 재정의된 var c
print(a,b,c,d, sep='\t')

# 변수의 값 교환
a, b = b, a
print(a, b, sep = '\t')

# 함수의 리턴에 튜플을 사용해 여러 값을 리턴하고 할당할 수 있다.
m, n = 97,  40
x, y = divmod(m, n)
print(x, y)
```
