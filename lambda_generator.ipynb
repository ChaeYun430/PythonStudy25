```python
# generator
# 함수 내부에 yield 키워드를 사용하면 해당 함수는 generator 함수가 되며,
# 일반 함수와는 달리 함수를 호출해도 함수 내부의 코드가 실행되지 않는다.
# next()를 사용해 실행하면 yield 키워드 부분까지만 실행하고, 리턴값으로 yield 키워드 뒤에 입력한 값이 출력된다.
# next() 호풀 후 yield 키워드를 만나지 못하고 함수가 끝나면 StopIteration 예외 발생
def test():
  print('A')
  yield 1
  print('B')
  yield 2
  print('C')
output = test()
print('D')
x = next(output)
print(x)
print('E')
y = next(output)
print(y)
print('F')
z = next(output)
print(z)
print('G')
last = next(output)
print(last)
```
